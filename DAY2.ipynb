```python
import random

position = 0

rabbit_list = (5, 19, 34, 45, 49, 53, 55, 58, 60, 65, 89, 91, 93, 95, 99)

print("🐰 Rabbit & Carrot Game Started 🥕")

while position < 100:
    input("Press Enter to roll the dice 🎲...")
    dice = random.randint(1, 6)
    print(f"\n🎲 Dice rolled: {dice}")

    position += dice

    # Boundary check
    if position > 100:
        position -= dice
        print("🚫 Can't move beyond 100!")

    # Rabbit trap
    if position in rabbit_list:
        print("🐰 Oh no! Rabbit trap!")
        position -= 10

    print(f"📍 Current Position: {position}")

print("\n🎉 Congratulations! Carrot reached home 🏆")

    

 
```
